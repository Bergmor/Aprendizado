```python
import csv
from matplotlib import pyplot as plt
import collections
```

```python
file='customers.csv'
```

```python
with open(file, 'rU') as infile:
    reader = csv.DictReader(infile)
    data = {}
    for row in reader:
      for header, value in row.items():
        try:
          data[header].append(value)
        except KeyError:
          data[header] = [value]

# extract the variables you want
customerId = data['CustomerID']
gender = data['Gender']
age = list(map(int, data['Age']))
annualIncome = data['AnnualIncome']
spendingScore = data['SpendingScore']
profession = data['Profession']
workExperience = data['WorkExperience']
familySize = data['FamilySize']
```

```python
counterGender = collections.Counter(gender)
gender_list = list(counterGender.keys())
counter_gender_list = list(counterGender.values())
```

```python
fig, ax = plt.subplots()

ax.pie(counter_gender_list, labels=gender_list, autopct='%.1f%%')

plt.show()
```

```python
counterAge = collections.OrderedDict(sorted(collections.Counter(age).items()))
age_list = list(counterAge.keys())
counter_age_list = list(counterAge.values())
```

```python
fig, ax = plt.subplots()

ax.bar(age_list, counter_age_list, color='blue', width = 0.3)
ax.set_xlabel('Idade')
ax.set_ylabel('Total')
ax.set_title('Total clientes por idade')

plt.show()
```

```python
counterProfession = collections.Counter(profession)
profession_list = list(counterProfession.keys())
counter_profession_list = list(counterProfession.values())
```

```python
fig, ax = plt.subplots()
ax.barh(profession_list, counter_profession_list)
for i, v in enumerate(counter_profession_list):
    ax.annotate(str(abs(v)), xy=(v, i), ha='left', va='center')
ax.set_xlabel('Quantidade profissionais')
ax.set_title('Total profissionais')
plt.show()
```
